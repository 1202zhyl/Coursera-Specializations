```python
f = open('dataset_3_2.txt', 'r')
```

```python
strand = f.read().strip()
```

```python
def complement(s):
    if s=='A': return 'T'
    if s=='G': return 'C'
    if s=='T': return 'A'
    return 'G'
```

```python
complementary = ''
for s in strand:
    complementary = complement(s) + complementary
```

```python
print('Complementary is: ' + complementary)
```

```python
import re # import regex library
```

```python
def find_indices(string, pattern):
    indices = [match.start() for match in re.finditer('(?=%s)' % pattern, string)] 
    f = open('output.txt', 'w')
    result = ''
    for index in indices:
        result += str(index) + ' '
    f.write(result)
```

```python
f = open('dataset_3_5.txt', 'r')
```

```python
arguments = f.readlines()
```

```python
string = arguments[1].strip()
pattern = arguments[0].strip()
```

```python
find_indices(string, pattern)
```
